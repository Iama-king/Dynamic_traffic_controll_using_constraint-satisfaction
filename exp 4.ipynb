```python
import random as r
```

```python
class junction:
    def __init__(self, lil=[],ril=[],uil=[],dil=[],ll=None, rl=None, ul=None ,dl=None):
        self.lil=lil
        self.ril=ril
        self.uil=uil
        self.dil=dil
        self.ll=ll
        self.rl=rl
        self.ul=ul
        self.dl=dl
```

```python
j1=junction()
j2=junction()
j3=junction()
j4=junction()
```

```python
tl=[[[1,2,3],[2,3,4],[1,1,1],[2,2,2]],[[1,2,3],[2,3,4],[1,1,1],[2,2,2]],[[1,2,3],[2,3,4],[1,1,1],[2,2,2]],[[1,2,3],[2,3,4],[1,1,1],[2,2,2]]]
```

```python
def inp(j,i):
    j.lil=tl[i][0]
    j.uil=tl[i][1]
    j.ril=tl[i][2]
    j.dil=tl[i][3]
    #j.lil=list(eval(input("enter your vehicles")))
    #j.ril=list(eval(input("enter your vehicles")))
    #j.uil=list(eval(input("enter your vehicles")))
    #j.dil=list(eval(input("enter your vehicles")))
    return j
        
```

```python
jl=[j1,j2,j3,j4]
for i in range(4):
    jl[i]=inp(jl[i],i)
```

```python
j1.rl=j2
j1.dl=j3
j2.ll=j1
j2.dl=j4
j3.ul=j1
j3.rl=j4
j4.ll=j3
j4.ul=j2
```

```python
def priot(j):
    r=j.ril[1]*6 + j.ril[1]*2 + j.ril[0]
    l=j.lil[1]*6 + j.lil[1]*2 + j.lil[0]
    u=j.uil[1]*6 + j.uil[1]*2 + j.uil[0]
    d=j.dil[1]*6 + j.dil[1]*2 + j.dil[0]
    l1=[r,l,u,d]
    p1=['r','l','u','d']
    for i in range(4):
        for j in range(i,4):
            if(l1[i]<l1[j]):
                l1[i],l1[j]=l1[j],l1[i]
                p1[i],p1[j]=p1[j],p1[i]
    return p1
```

```python
def prioj(j):
    r=j.ril[1]*6 + j.ril[1]*2 + j.ril[0]
    l=j.lil[1]*6 + j.lil[1]*2 + j.lil[0]
    u=j.uil[1]*6 + j.uil[1]*2 + j.uil[0]
    d=j.dil[1]*6 + j.dil[1]*2 + j.dil[0]
    l1=[r,l,u,d]
    return sum(l1) 
```

```python
def tc(j):
    dic={'r':j.ril,'l':j.lil,'u':j.uil,'d':j.dil}    
    pl=priot(j)
    print("priority",pl)
    for i in pl:
        lq=len(dic[i])
        for o in range (lq):
            for k in range(dic[i][o]):
                di=r.randint(0,4)
                if(di==0):
                    if(j.ul!=None):
                        j.ul.dil[o]+=1
                if(di==1):
                    if(j.rl!=None):
                        j.rl.lil[o]+=1
                if(di==2):
                    if(j.dl!=None):
                        j.dl.uil[o]+=1
                if(di==3):
                    if(j.ll!=None):
                        j.ll.ril[o]+=1
    j.ril=[0,0,0]
    j.lil=[0,0,0]
    j.uil=[0,0,0]
    j.dil=[0,0,0]
    
```

```python
f=open("csa.txt","a")
def jpf():
    zc=0
    jpl=[]
    for i in jl:
        si=prioj(i)
        jpl.append(si)
        if(si==0):
            zc+=1
    if(zc==4):
        return "end"
    m=0
    ind=0
    for i in range(4):
        if(jpl[i]>m):
            m=jpl[i]
            ind=i
    con(j1,j2,j3,j4)
    return ind
```

```python
while(True):
    v=jpf()
    print("junction pro",v)
    if(v=="end"):
        print("end")
        break
    else:
        tc(jl[v])
```
